## Reemplazo con moda

```python
import matplotlib.pyplot as plt
def plot_cols(df, cols):
  fig, plots = plt.subplots(nrows=1, ncols=len(cols), figsize=(20,4))
  for i, col in enumerate(cols):
    columna= df[col].astype(str)
    ax = plots.flat[i] if len(cols) > 1 else plots
    ax.hist(columna, bins=20)
    ax.set_title(col)
    ax.set_ylabel("frecuencia")
    ax.set_xticks(ax.get_xticks())
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
  fig.show()
```

```python
plot_cols(df, ["ESTU_VALORMATRICULAUNIVERSIDAD", "ESTU_HORASSEMANATRABAJA", "FAMI_TIENEINTERNET", "ESTU_PAGOMATRICULAPROPIO"])
```

```python
def fillna_mode(df, cols):
  for col in cols:
    moda = df[col].mode().values[0]
    df.fillna({col: moda}, inplace=True)
# en todos los casos como se trata de muy pocos datos nulos en comparación con la moda se opta por aplicarlos a la moda
fillna_mode(df, ["ESTU_VALORMATRICULAUNIVERSIDAD", "ESTU_HORASSEMANATRABAJA", "FAMI_TIENEINTERNET", "ESTU_PAGOMATRICULAPROPIO"])
plot_cols(df, ["ESTU_VALORMATRICULAUNIVERSIDAD", "ESTU_HORASSEMANATRABAJA", "FAMI_TIENEINTERNET", "ESTU_PAGOMATRICULAPROPIO"])
```

```python
def fillna_proportional(df, cols):
  for col in cols:
    # se toman las dos primeras modas
    moda1, moda2 = df[col].value_counts()[:2]
    m1_p = moda1 / (moda1 + moda2) # proporción moda1
    na_count = df[col].isna().sum()
    valor_moda1, valor_moda2 = df[col].value_counts().index[:2]
    df.fillna({col: valor_moda1}, inplace=True, limit=int(m1_p*na_count))
    df.fillna({col: valor_moda2}, inplace=True)
    assert df[col].isna().sum() == 0

fillna_proportional(df, ["FAMI_ESTRATOVIVIENDA", "FAMI_EDUCACIONPADRE", "FAMI_EDUCACIONMADRE"])
plot_cols(df, ["FAMI_ESTRATOVIVIENDA", "FAMI_EDUCACIONPADRE", "FAMI_EDUCACIONMADRE"])
```

```python
import unicodedata
from tqdm import tqdm
from spellchecker import SpellChecker

spell = SpellChecker(language='es')
new_words = ["artes"] # palabras que pyspellchecker no conoce
spell.word_frequency.load_words(new_words)
ignored_words = ["administración", "profesional", "licenciatura", "universitario"]

if "already_corrected" not in locals():
  already_corrected = {}

# Función para eliminar acentos
def normalize(input_str):
  nfkd_form = unicodedata.normalize('NFKD', input_str)
  return "".join([c for c in nfkd_form if not unicodedata.combining(c)]).lower().strip() if pd.notnull(input_str) else input_str

def correct_words(input_str):
  if pd.isnull(input_str):
    return input_str
  corrected_words = []
  for word in input_str.split():
    if word in already_corrected:
      # Usar la corrección previamente almacenada
      corrected_words.append(already_corrected[word])
    else:
      # Realizar la corrección y almacenarla en la memoria
      corrected_word = spell.correction(word) if spell.correction(word) else word
      already_corrected[word] = corrected_word
      corrected_words.append(corrected_word)
  return ' '.join(corrected_words)

def remove_redundant_words(input_str):
  return " ".join([w for w in input_str.split() if w not in ignored_words])

def clean_programs(df):
  cleaned_df = df.copy()

  # Eliminar múltiples espacios
  cleaned_df['ESTU_PRGM_ACADEMICO'] = cleaned_df['ESTU_PRGM_ACADEMICO'].str.replace(r'\s+', ' ', regex=True)

  # Aplicar transformación: eliminar acentos, convertir a minúsculas y eliminar espacios extras
  tqdm.pandas(desc="Normalizando texto")
  cleaned_df['ESTU_PRGM_ACADEMICO'] = cleaned_df['ESTU_PRGM_ACADEMICO'].progress_apply(normalize)

  tqdm.pandas(desc="Corrigiendo palabras")
  cleaned_df["ESTU_PRGM_ACADEMICO"] = cleaned_df["ESTU_PRGM_ACADEMICO"].progress_apply(correct_words)

  # Aquí se eliminan palabras que no aportan mucho al algoritmo de agrupamiento
  # tqdm.pandas(desc="Eliminando palabras redundates")
  # cleaned_df["ESTU_PRGM_ACADEMICO"] = cleaned_df["ESTU_PRGM_ACADEMICO"].progress_apply(remove_redundant_words)

  return cleaned_df
```

## Agrupamiento de programas similares utilizando DBSCAN

```python
import numpy as np
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import DBSCAN

nltk.download('stopwords')
stop_words_spanish = stopwords.words('spanish')
stop_words_spanish = stop_words_spanish + ignored_words

sens_params = {
  "min_df": [4, 2],
  "eps": [0.6, 0.7],
  "min_samples": [10, 2]
}

def group_programs(programs, sens=0):
  min_df = sens_params["min_df"][sens]
  eps = sens_params["eps"][sens]
  min_samples = sens_params["min_samples"][sens]

  # Crear una representación numérica usando TF-IDF
  vectorizer = TfidfVectorizer(stop_words=stop_words_spanish, min_df=min_df, ngram_range=(1, 2))

  # Crear el TF-IDF transformando todas las filas
  programs_df = programs.copy().to_frame()
  Xv = vectorizer.fit_transform(programs_df['ESTU_PRGM_ACADEMICO'].dropna())

  # Inicializar DBSCAN
  db = DBSCAN(eps=eps, min_samples=min_samples, metric='cosine')
  batch_size = 5000
  n_batches = Xv.shape[0] // batch_size + 1

  # resultados de los clusters
  clusters = np.full(Xv.shape[0], -1)  # Iniciar todos los clusters con -1 (no asignado)

  # Iterar sobre los lotes y aplicar DBSCAN
  for batch_start in tqdm(range(0, Xv.shape[0], batch_size), desc="Aplicando DBSCAN"):
      batch_end = min(batch_start + batch_size, Xv.shape[0])  # Definir el final del lote
      batch_data = Xv[batch_start:batch_end]  # Obtener el lote de datos

      # Aplicar DBSCAN en el lote
      cluster_batch = db.fit_predict(batch_data)

      # Asignar los resultados a los clusters correspondientes
      clusters[batch_start:batch_end] = cluster_batch

  # Asignar los resultados de los clusters al DataFrame
  programs_df['cluster'] = clusters
  return programs_df

programs_df = group_programs(xx["ESTU_PRGM_ACADEMICO"])
print("\nNúmero total de clusters: ", programs_df["cluster"].max() + 1) # contando el 0 y sin contar el -1
```

# **Entrenamiento**

```python
X_t = xx.copy()[:nx]
y_t = X_t["RENDIMIENTO_GLOBAL"][:nx]
X_t = X_t.drop(["ID", "RENDIMIENTO_GLOBAL"], axis=1) # descartamos columnas que no aportan nada o son la columna a predecir
X_t.head()
```

## Uso de red neuronal

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Input, Dropout, BatchNormalization
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X_t, y_t, test_size=0.2, random_state=4)

# Normalizar las características
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print(X_train)

# Crear un modelo secuencial
model = Sequential()

model.add(Input((X_train.shape[1],)))

# model.add(Dense(512, activation='relu'))
# model.add(BatchNormalization())
# model.add(Dropout(0.5))

model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))

model.add(Dense(128, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.25))

model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
# model.add(Dropout(0.5))

# Capa de salida (número de clases para 'RENDIMIENTO_GLOBAL')
model.add(Dense(4, activation='softmax'))  # 4 clases (bajo, medio-bajo, medio-alto, alto)

# Compilar el modelo
model.compile(loss='sparse_categorical_crossentropy',  # Pérdida para clasificación múltiple
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# Entrenar el modelo
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
early = EarlyStopping(monitor='val_loss', patience=8)
reducelr = ReduceLROnPlateau(monitor='val_loss', patience=15, min_lr=0.001)
history = model.fit(X_train, y_train, epochs=128, batch_size=64, validation_data=(X_test, y_test), callbacks=[early, reducelr])
```

```python
# Evaluar el modelo
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Loss: {loss}")
print(f"Accuracy: {accuracy}")
```
