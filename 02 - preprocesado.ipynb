# **Datos nulos Repartidos proporcionalmente**

```python
columna= x["FAMI_ESTRATOVIVIENDA"].astype(str)
plt.hist(columna, bins=20)
plt.title("histograma FAMI_ESTRATOVIVIENDA")
plt.ylabel("frecuencia")
plt.xlabel("FAMI_ESTRATOVIVIENDA")
plt.xticks(rotation=90)
plt.show()
```

```python
print(x["FAMI_ESTRATOVIVIENDA"].value_counts()["Estrato 2"])
print(x["FAMI_ESTRATOVIVIENDA"].value_counts()["Estrato 3"])
print(x["FAMI_ESTRATOVIVIENDA"].isna().sum())
```

```python
# en este caso tenemos que el estrato dos es la moda, sin embargo el estrato 3 no se aleja mucho, por lo que se decide repartir los datos nulos entre ellas proporcionalmente
col= x["FAMI_ESTRATOVIVIENDA"]
f2= col.value_counts()["Estrato 2"]
f3= col.value_counts()["Estrato 3"]
fr = f2 / f3
nulos = col.isna().sum()
vn2 = int(nulos/fr)
vn3= int(nulos-vn2)
x["FAMI_ESTRATOVIVIENDA"].fillna("Estrato 2", inplace=True, limit= vn2)
x["FAMI_ESTRATOVIVIENDA"].fillna("Estrato 3", inplace=True, limit= vn3)
```

```python
#en este caso tenemos bastantes datos nulos y moda no lejana, se distribuyen entre los dos mas grandes
print(x["FAMI_EDUCACIONPADRE"].value_counts()["Secundaria (Bachillerato) completa"])
print(x["FAMI_EDUCACIONPADRE"].value_counts()["Primaria incompleta"])
print(x["FAMI_EDUCACIONPADRE"].isna().sum())
#----------------------------------

col= x["FAMI_EDUCACIONPADRE"]
f2= col.value_counts()["Secundaria (Bachillerato) completa"]
f3= col.value_counts()["Primaria incompleta"]
fr = f2 / f3
nulos = col.isna().sum()
vn2 = int(nulos/fr)
vn3= int(nulos-vn2)
x["FAMI_EDUCACIONPADRE"].fillna("Secundaria (Bachillerato) completa", inplace=True, limit= vn2)
x["FAMI_EDUCACIONPADRE"].fillna("Primaria incompleta", inplace=True, limit= vn3)
#--------------------
print(x["FAMI_EDUCACIONPADRE"].value_counts()["Secundaria (Bachillerato) completa"])
print(x["FAMI_EDUCACIONPADRE"].value_counts()["Primaria incompleta"])
print(x["FAMI_EDUCACIONPADRE"].isna().sum())
```
